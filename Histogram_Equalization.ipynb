```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread('/content/image.jpg',0)

equalized_img = cv2.equalizeHist(img)

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(equalized_img,cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")
```
